```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegressionCV
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
df = pd.read_excel("crop yield data sheet.xlsx")
df
```

## EDA

```python
df = df.dropna()
df
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df['Temperatue'].replace({np.nan:23},inplace =True)
```

```python
df.info()
```

```python
scaler = StandardScaler().fit_transform(df)
df_scaled = pd.DataFrame(scaler, columns = df.columns)
df_scaled
```

```python
x = df_scaled.drop("Yeild (Q/acre)", axis = 1)
y = df_scaled["Yeild (Q/acre)"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state=  42)
```

```python
dt_clf = DecisionTreeRegressor(random_state=42)
dt_clf.fit(x_train,y_train)
```

```python
dt_clf.score(x_train,y_train)
```

```python
dt_clf.score(x_test,y_test)
```

## Hyperparameter Tuning

```python
dt_model = DecisionTreeRegressor(random_state=42)

hyp_grid = {
            "criterion" :['squared_error'],
            
            'max_depth':np.arange(2,5),
            'min_samples_split': range(3,8),
            'min_samples_leaf': range(10,15)}

gscv_dt_model = GridSearchCV(dt_model,param_grid = hyp_grid, cv = 3,n_jobs  = -1)
gscv_dt_model.fit(x_train, y_train)
```

```python
dt_clf = gscv_dt_model.best_estimator_
dt_clf.fit(x_train,y_train)
print("Train accuracy ;",dt_clf.score(x_train,y_train))
print("Testing accuracy :",dt_clf.score(x_test,y_test))
plt.figure(figsize = (100,90))
tree = plot_tree(dt_clf, feature_names=x.columns, filled=True)
plt.savefig('DT_clf_test.png')
```

## Training data evaluation


```python
dt_clf.score(x_train,y_train)
```

```python
dt_clf.score(x_test,y_test)
```

```python
y_pred_train = dt_clf.predict(x_train)

mse = mean_squared_error(y_pred_train,y_train)
print("mean squared error :",mse)
print("*"*50)

mae = mean_absolute_error(y_pred_train,y_train)
print("mean absolute error :",mae)
print("*"*50)
r_score = r2_score(y_pred_train,y_train)
print("r2_score :",r_score)
```

## testing_data_evaluation

```python
y_pred = dt_clf.predict(x_test)

mse = mean_squared_error(y_pred,y_test)
print("mean squared error :",mse)
print("*"*50)

mae = mean_absolute_error(y_pred,y_test)
print("mean absolute error :",mae)
print("*"*50)
r_score = r2_score(y_pred,y_test)
print("r2_score :",r_score)
```
